```python
import networkx as nx
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
#create an empty graph with no nodes and no edges

g = nx.Graph()
```

```python
#add a single node

g.add_node(1)
```

```python
#add a list of nodes

g.add_nodes_from([2,3])
```

```python
nx.draw_networkx(g, note_color='green', node_size=700)
```

```python
#adding edges using unpack edge tupels
g.add_edge(1,2)

nx.draw_networkx(g)
```

```python
#adding edges using unpack edge tupels
e = (2,3)
g.add_edge(*e)

nx.draw_networkx(g)
```

```python
#adding list of edges
g.add_edges_from([(1,2), (1,3)])

nx.draw_networkx(g)
```

```python
#get some properties in graph

#find number of a nodes in a graph
print(g.number_of_nodes())
```

```python
#find number of edges in a graph
print(g.number_of_edges())
```

```python
#getting a number of nodes in graph
print(g.nodes())
```

```python
#getting a number of edges in graph
print(g.edges())
```

```python
#getting a neighbor from connected to the graph
print(g.neighbors(1))
```

```python
#get a degree from a graph
print(g.degree())
```

```python
#find if a graph has an edges
print(g.has_edge(1,3))
```

```python
#find if there exist a node in a graph
print(g.has_node(3))
```

```python
#creating a directed graph
b = nx.DiGraph([('a','b'), ('b','c'), ('c', 'a')])
```

```python
#adding edges to the di-graph
b.add_edge('a', 'b')
b.add_edges_from([('d', 'c'), ('d', 'b')])
```

```python
#plot our graph
nx.draw_networkx(b, with_labels=True, node_size=800, node_colot='cyan')
```

```python
#print the out degrees of the di-graph
print(b.out_degree())
```

```python
#print the in degrees of the di-graph
print(b.in_degree())
```

```python
#print some other properties in the graph
print(b.in_edges())
```

```python
print(b.out_edges())
```
