```python
import networkx as nx
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
#create an empty graph with no nodes and no edges

g = nx.Graph()
```

```python
#add a single node

g.add_node(1)
```

```python
#add a list of nodes

g.add_nodes_from([2,3])
```

```python
nx.draw_networkx(g, note_color='green', node_size=700)
```

```python
#adding edges using unpack edge tupels
g.add_edge(1,2)

nx.draw_networkx(g)
```

```python
#adding edges using unpack edge tupels
e = (2,3)
g.add_edge(*e)

nx.draw_networkx(g)
```
